```python
from pydrake.all import (
    DiagramBuilder,
    Simulator,
    LeafSystem,
)
from realsense_hardware import PlanarPositionDetector
```

```python
class Printer(LeafSystem):
    def __init__(self, size):
        super().__init__()
        self.DeclareVectorInputPort("data", size)
        self.DeclarePerStepPublishEvent(self.Publish)

    def Publish(self, context):
        data = self.get_input_port().Eval(context)
        print(data)
```

```python
builder = DiagramBuilder()

position_detector = builder.AddSystem(PlanarPositionDetector())
printer = builder.AddSystem(Printer(3))
builder.Connect(position_detector.get_output_port(), printer.get_input_port())

diagram = builder.Build()
simulator = Simulator(diagram)
simulator.set_target_realtime_rate(1.0)

simulator.AdvanceTo(1000)
```

```python
position_detector.Close()
```
